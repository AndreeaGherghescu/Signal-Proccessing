```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import scipy.io.wavfile
```

```python
# Exercitiul 1

# axa de numere
axa = np.arange(0, 0.03, 0.0005)
print(axa)
```

```python
# functiile date in cerinta, unde t este timpul

def X(t):
    # x(t) = cos(520πt + π/3)
    return np.cos(520 * np.pi * t + np.pi / 3)

def Y(t):
    # y(t) = cos(280πt − π/3)
    return np.cos(280 * np.pi * t - np.pi / 3)

def Z(t):
    # z(t) = cos(120πt + π/3)
    return np.cos(120 * np.pi * t - np.pi / 3)
```

```python
# construim graficele
fig, axs = plt.subplots(3)
fig.suptitle("Punctul 1b")

# aplicam functiile la fiecare moment de timp pentru a vedea valoarea semnalului
list_x = list(map(X, axa))
list_y = list(map(Y, axa))
list_z = list(map(Z, axa))

# subploturile
axs[0].set_title('x(t)')
axs[0].plot(axa, list_x)
axs[1].set_title('y(t)')
axs[1].plot(axa, list_y)
axs[2].set_title('z(t)')
axs[2].plot(axa, list_z)

# las spatiu intre subploturi
plt.subplots_adjust(hspace=1)

# adaug linii de grid in subploturi
for subplot in axs:
    subplot.grid(True)

plt.show()
```

```python
# 1 c

# frecventa de esantionare
# fs = 200 Hz

fs = 200
time = 0.03
step = 1 / fs
# axa de numere
axa_c = np.arange(0, time, step)

# semnalele esantionate
def X_esantion(t):
    # x(t) = cos(520πt + π/3)
    return np.cos(520 * np.pi * t + np.pi / 3)

def Y_esantion(t):
    # y(t) = cos(280πt − π/3)
    return np.cos(280 * np.pi * t - np.pi / 3)

def Z_esantion(t):
    # z(t) = cos(120πt + π/3)
    return np.cos(120 * np.pi * t - np.pi / 3)

# construim graficele
fig_c, axs_c = plt.subplots(3)
fig_c.suptitle("Punctul 1c")

# aplicam functiile la fiecare moment de timp pentru a vedea valoarea semnalului
es_x = X_esantion(axa_c)
es_y = Y_esantion(axa_c)
es_z = Z_esantion(axa_c)

# subploturile
axs_c[0].set_title('x(t)')
axs_c[0].plot(axa_c, es_x)
axs_c[1].set_title('y(t)')
axs_c[1].plot(axa_c, es_y)
axs_c[2].set_title('z(t)')
axs_c[2].plot(axa_c, es_z)

# las spatiu intre subploturi
plt.subplots_adjust(hspace=1)

# adaug linii de grid in subploturi
for subplot in axs_c:
    subplot.grid(True)

plt.show()
```

```python
# Exercitiul 2
# a
# 400 Hz, 1600 esantioane

axa_2a = np.linspace(0, 1, 30000)
# axa_2a = np.linspace(0, 0.03, 1600)

f0 = 400
# semnalul
def sin_a (t):
    return np.sin(2 * np.pi * f0 * t)

# construiesc graficul
fig_2a, axs_2a = plt.subplots()
fig_2a.suptitle("Punctul 2a")

# calculez valorile
x_2a = list(map(sin_a, axa_2a))

# afisez
axs_2a.plot(axa_2a, x_2a)
axs_2a.grid(True)
plt.show()


# Lab 2
rate = int(10e5)
x_2a = np.array(x_2a)
scipy.io.wavfile.write('2a.wav', rate, x_2a)
```

```python
# 2 b
# 800 Hz, 3 sec

axa_2b = np.linspace(0, 3, 800)

f0 = 800
# semnalul
def sin_b (t):
    return np.sin(2 * np.pi * f0 * t)

# construiesc graficul
fig_2b, axs_2b = plt.subplots()
fig_2b.suptitle("Punctul b")

# calculez valorile
x_2b = list(map(sin_b, axa_2b))

# afisez
axs_2b.plot(axa_2b, x_2b)
axs_2b.grid(True)
plt.show()

# Lab 2
rate = int(10e5)
x_2b = np.array(x_2b)
scipy.io.wavfile.write('2b.wav', rate, x_2b)
```

```python
# 2 c
# sawtooth, 240 Hz

axa_2c = np.linspace(0, 0.03, 160)

f0 = 240
# semnalul
def sawtooth (t):
    return f0 * t - np.floor(f0 * t)

# construiesc graficul
fig_2c, axs_2c = plt.subplots()
fig_2c.suptitle("Punctul 2c")

# calculez valorile
x_2c = list(map(sawtooth, axa_2c))

# afisez
axs_2c.plot(axa_2c, x_2c)
axs_2c.grid(True)
plt.show()

# Lab 2
rate = int(10e5)
x_2c = np.array(x_2c)
scipy.io.wavfile.write('2c.wav', rate, x_2c)
```

```python
# 2 d
# square, f0 = 300 Hz

axa_2d = np.linspace(0, 0.03, 100)
f0 = 300

def square (t):
    return signal.square(2 * np.pi * f0 * t)

# construiesc graficul
fig_2d, axs_2d = plt.subplots()
fig_2d.suptitle("Punctul 2d")

# calculez valorile
x_2d = list(map(square, axa_2d))

# afisez
axs_2d.plot(axa_2d, x_2d)
axs_2d.grid(True)
plt.show()

# Lab 2
rate = int(10e5)
x_2d = np.array(x_2d)
scipy.io.wavfile.write('2d.wav', rate, x_2d)
```

```python
# Exercitiul 3
# fs = 2000 Hz
fs = 2000

# a
# fs reprezinta numarul de esantioane pe secunda
# fs = 2 * pi / T, T este perioada de esantionare
# T = 2 * pi / fs
T = 2 * np.pi / 2000
print(f"{T} secunde intre esantioane")

# b
# un esantion -> 4 biti
# cati bytes pentru 1h

# t = 1h = 60 min = 3600 sec
t = 3600
# 2000 esantioane .... 1 sec
# x .................. 3600 sec

x = fs * t # esantioane

# 1 byte -> 8 biti

# 1 esantion .... 4 biti
# x esantioane .. y biti

y = x * 4  # biti

print(f"{y/8} bytes")
```
