```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Exercitiul 1

# axa de numere
# frecv esantionare < 2 * f0

f0 = 4
f1 = 11
f2 = 18

fs2 = 7
# fs - frecv de esantionare = 7 => 8 esantioane

axa = np.arange(0, 1, 0.00001)
axa2 = np.linspace(0, 1, fs2 + 1)

# semnal de tip sinus
def sin (t, f):
    return np.sin(2 * np.pi * f * t)

# construim graficele
fig, axs = plt.subplots(4)
fig.suptitle("Exercitiul 1")

# aplicam functiile la fiecare moment de timp pentru a vedea valoarea semnalului
val = sin(axa, f0)
val1 = sin(axa, f1)
val2 = sin(axa, f2)

# subploturile
axs[0].set_title('functia 1')
axs[0].plot(axa, val)
axs[1].set_title('functia 1')
axs[1].plot(axa, val)
axs[1].scatter(axa2, sin(axa2, f0), color = 'y')
axs[2].set_title('functia 2')
axs[2].plot(axa, val1)
axs[2].scatter(axa2, sin(axa2, f1), color = 'y')
axs[3].set_title('functia 3')
axs[3].plot(axa, val2)
axs[3].scatter(axa2, sin(axa2, f2), color = 'y')

# las spatiu intre subploturi
plt.subplots_adjust(hspace=1)

# adaug linii de grid in subploturi
for subplot in axs:
    subplot.grid(True)

plt.show()
```
