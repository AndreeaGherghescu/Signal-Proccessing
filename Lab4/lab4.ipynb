```python
# c
axa = np.arange(12 * 60 * 60, 7 * 24 * 60 * 60, 2 * 60 * 60)
axa2 = np.arange(12 * 60 * 60, 7 * 24 * 60 * 60, 2)

def signal(t):
    return np.sin(2 * np.pi * f0 * t - np.pi / 2)

# construim graficul
fig, axs = plt.subplots(1)
fig.suptitle("Semnal")

# aplicam functia la fiecare moment de timp pentru a vedea valoarea semnalului
values = signal(axa2)
points = signal(axa)

# subplotul
axs.set_title('Semnal')
axs.plot(axa2, values)
axs.scatter(axa, points, color = 'y')

plt.show()
```

```python
# Exercitiul 2

# axa de numere
axa = np.arange(0, 0.03, 0.00005)

f0 = 400

def sin(t):
    return 2 * np.sin(2 * np.pi * f0 * t)

# construim graficele
fig, axs = plt.subplots(1)
fig.suptitle("Exercitiul 2")

# aplicam functia la fiecare moment de timp pentru a vedea valoarea semnalului
valSin = sin(axa)
# shiftam functia mai sus
valSin = valSin + 5

# afisam
axs.plot(axa, valSin)

plt.show()
```

```python
# Exercitiul 3

B = 10
fc = 90

def calcFOptim (m):
    lower = (2 * fc + B) / (m + 1)
    higher = (2 * fc - B) / m
    return lower, higher

# a
m = 1
print(f"Frecventele optime: {calcFOptim(m)}")
print("Optimul este 95.0, deoarece scazand aceasta valoare replica va fi lipita la stanga de 0.")
print()
# b
m = 2
print(f"Frecventele optime: {calcFOptim(m)}")
print("Optimul este 85.0, deoarece scazand aceasta valoare replica va fi lipita la dreapta de 0.")
print()
# c
m = 4
print(f"Frecventele optime: {calcFOptim(m)}")
print("Optimul este 42.5, deoarece replicile vor fi la distanta fs una de alta, deci una  din ele va fi lipita de 0 (85 = 42.5 * 2).")
print()

```
