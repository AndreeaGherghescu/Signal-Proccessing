```python
# Exercitiul 1

# axa de numere
axa = np.arange(0, 0.03, 0.0005)

f0 = 400

# semnal de tip sinus
def sin (t):
    return np.sin(2 * np.pi * f0 * t + np.pi / 3)

def cos (t):
    return np.cos(2 * np.pi * f0 * t + np.pi / 3 - np.pi / 2)

# construim graficele
fig, axs = plt.subplots(2)
fig.suptitle("Exercitiul 1")

# aplicam functiile la fiecare moment de timp pentru a vedea valoarea semnalului
valSin = sin(axa)
valCos = cos(axa)

# subploturile
axs[0].set_title('sin(t)')
axs[0].plot(axa, valSin)
axs[1].set_title('cos(t)')
axs[1].plot(axa, valCos)

# las spatiu intre subploturi
plt.subplots_adjust(hspace=1)

# adaug linii de grid in subploturi
for subplot in axs:
    subplot.grid(True)

plt.show()

```

```python
# Exercitiul 2

# axa de numere
axa = np.arange(0, 0.03, 0.0005)

f0 = 400
phase = [np.pi / 3, 3 * np.pi / 4, np.pi / 2, np.pi]

# semnal de tip sinus
def sin (t, i):
    return np.sin(2 * np.pi * f0 * t + phase[i])

# construim graficele
fig, axs = plt.subplots(4)
fig.suptitle("Exercitiul 2")

# aplicam functiile la fiecare moment de timp pentru a vedea valoarea semnalului
val0 = sin(axa, 0)
val1 = sin(axa, 1)
val2 = sin(axa, 2)
val3 = sin(axa, 3)

# subploturile
axs[0].set_title('phase 0')
axs[0].plot(axa, val0)
axs[1].set_title('phase 1')
axs[1].plot(axa, val1)
axs[2].set_title('phase 2')
axs[2].plot(axa, val2)
axs[3].set_title('phase 3')
axs[3].plot(axa, val3)

# las spatiu intre subploturi
plt.subplots_adjust(hspace=1)

# adaug linii de grid in subploturi
for subplot in axs:
    subplot.grid(True)

plt.show()

```

```python
# Exercitiul 4

# axa de numere
axa = np.arange(0, 0.03, 0.0005)

f0 = 400

# semnal de tip sinus
def sin (t):
    return np.sin(2 * np.pi * f0 * t + np.pi / 3)

def sawtooth (t):
    return f0 * t - np.floor(f0 * t)

# construim graficele
fig, axs = plt.subplots(3)
fig.suptitle("Exercitiul 4")

# aplicam functiile la fiecare moment de timp pentru a vedea valoarea semnalului
valSin = sin(axa)
valSaw = sawtooth(axa)
valSum = valSaw + valSin

# subploturile
axs[0].set_title('sin(t)')
axs[0].plot(axa, valSin)
axs[1].set_title('sawtooth(t)')
axs[1].plot(axa, valSaw)
axs[2].set_title('sum(t)')
axs[2].plot(axa, valSum)

# las spatiu intre subploturi
plt.subplots_adjust(hspace=1)

# adaug linii de grid in subploturi
for subplot in axs:
    subplot.grid(True)

plt.show()
```

```python
# Exercitiul 5

# axa de numere
axa = np.linspace(0, 3, 3 * 44100)

f0 = 400
f1 = 800

# semnal de tip sinus
def sin (t, f):
    return np.sin(2 * np.pi * f * t + np.pi / 3)

# aplicam functiile la fiecare moment de timp pentru a vedea valoarea semnalului
val1 = sin(axa, f0)
val2 = sin(axa, f1)


# pun al doilea semnal la sfarsitul primului
valSuma = np.append(val1, val2)

# salvez semnalul final
rate = int(44100)
scipy.io.wavfile.write('5.wav', rate, valSuma)

# al doilea semnal, avand frecventa mai mare, este mai ascutit
```

```python
# Exercitiul 6

# axa de numere
axa = np.linspace(0, 3, 3 * 44100)

fs = 44100
f0 = fs / 2
f1 = fs / 4
f2 = 0

# semnal de tip sinus
def sin (t, f):
    return np.sin(2 * np.pi * f * t)

# construim graficele
fig, axs = plt.subplots(3)
fig.suptitle("Exercitiul 6")

# aplicam functiile la fiecare moment de timp pentru a vedea valoarea semnalului
val1 = sin(axa, f0)
val2 = sin(axa, f1)
val3 = sin(axa, f2)

# subploturile
axs[0].set_title('frecventa 1')
axs[0].plot(axa, val1)
axs[1].set_title('frecventa 2')
axs[1].plot(axa, val2)
axs[1].set_title('frecventa 3')
axs[1].plot(axa, val3)

# las spatiu intre subploturi
plt.subplots_adjust(hspace=1)

# adaug linii de grid in subploturi
for subplot in axs:
    subplot.grid(True)

plt.show()

# pun al doilea semnal la sfarsitul primului
valSuma = np.append(val1, val2)

# salvez semnalul final
rate = int(44100)
scipy.io.wavfile.write('6a.wav', rate, val1)
scipy.io.wavfile.write('6b.wav', rate, val2)
scipy.io.wavfile.write('6c.wav', rate, val3)

# semnalul 2 se aude mult mai tare decat semnalul 1
# semnalul 3 are frecventa 0, deci nu auzim nimic
```

```python
# Exercitiul 7

# axa de numere
axa = np.linspace(0, 1, 4100)

f0 = 1000

# semnal de tip sinus
def sin (t):
    return np.sin(2 * np.pi * f0 * t)

# construim graficele
fig, axs = plt.subplots(3)
fig.suptitle("Exercitiul 7")

# aplicam functiile la fiecare moment de timp pentru a vedea valoarea semnalului

val = sin(axa)
mask = [True if i % 4 == 0 else False for i in range(len(val))]
axa1 = axa[mask]
val1 = sin(axa1)
mask1 = [True if i % 4 == 1 else False for i in range(len(val))]
axa2 = axa[mask1]
val2 = sin(axa2)


# subploturile
axs[0].set_title('Initial')
axs[0].plot(axa, val)
axs[1].set_title('Decimare de la indexul 0')
axs[1].plot(axa1, val1)
axs[2].set_title('Decimare de la indexul 1')
axs[2].plot(axa2, val2)

# las spatiu intre subploturi
plt.subplots_adjust(hspace=1)

# adaug linii de grid in subploturi
for subplot in axs:
    subplot.grid(True)

plt.show()

# a) graficul semnalului initial, fata de cel decimat, este mult mai dens
# b) decimarea este invarianta in timp
```

```python
# Exercitiul 8

# axa de numere
axa = np.linspace(-np.pi / 2, np.pi / 2, 1600)

def sin(a):
    return np.sin(a)

def a(a):
    return a

# construim graficele
fig, axs = plt.subplots(2)
fig.suptitle("Exercitiul 8")


valSin = sin(axa)
valA = a(axa)

# subploturile
axs[0].set_title('sin(a)')
axs[0].plot(axa, valSin)
axs[1].set_title('a')
axs[1].plot(axa, valA)

# las spatiu intre subploturi
plt.subplots_adjust(hspace=1)

# adaug linii de grid in subploturi
for subplot in axs:
    subplot.grid(True)

plt.show()

# Concluzie: aproximarea nu este tocmai corecta
```
